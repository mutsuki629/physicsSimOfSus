```python
from control import matlab
from matplotlib import pyplot as plt
from scipy import arange
```

```python
# 伝達関数の分子・分母多項式を与える
num = [0, 1] # 分子多項式
den1 = [3, 1] # T=3の場合の分母多項式
den2 = [8, 1] # T=8
den3 = [15, 1] # T=15

# 伝達関数表現を与える
sys1 = matlab.tf(num, den1) # T=3の場合の伝達関数表現
sys2 = matlab.tf(num, den2) # T=8
sys3 = matlab.tf(num, den3) # T=15
```

```python
# 時間変数の定義
t = arange(0, 50, 0.01) #0から10まで0.01刻み

# ステップ応答の計算
y1, t1 = matlab.step(sys1, t) # T=3の場合のステップ応答
y2, t2 = matlab.step(sys2, t)
y3, t3 = matlab.step(sys3, t)

# プロット
plt.plot(t1, y1, label="y_a(t)")
plt.plot(t2, y2, label="y_b(t)")
plt.plot(t3, y3, label="y_c(t)")
plt.xlim([0, 50])
plt.ylim([0, 1.2])
plt.xticks([0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50])
plt.grid(color='gray')
plt.xlabel('time t[s]')
plt.ylabel('y(t)')
plt.legend()
plt.show()
```
