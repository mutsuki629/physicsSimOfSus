```python
import numpy as np
from control import matlab
from matplotlib import pyplot as plt
from scipy import arange
```

```python
K = 1
zeta = 0.1
omegan1 = 1.0
omegan2 = 5.0
omegan3 = 10.0

num1 = [0, 0, K*omegan1**2]
den1 = [1, 2*zeta*omegan1, omegan1**2]
num2 = [0, 0, K*omegan2**2]
den2 = [1, 2*zeta*omegan2, omegan2**2]
num3 = [0, 0, K*omegan3**2]
den3 = [1, 2*zeta*omegan3, omegan3**2]

sys1 = matlab.tf(num1, den1)
sys2 = matlab.tf(num2, den2)
sys3 = matlab.tf(num3, den3)
```

```python
t = arange(0, 15, 0.01)

y1, t1 = matlab.impulse(sys1, t)
y2, t2 = matlab.impulse(sys2, t)
y3, t3 = matlab.impulse(sys3, t)

plt.plot(t1, y1, label="omegan_n = 1")
plt.plot(t2, y2, label="omegan_n = 5")
plt.plot(t3, y3, label="omegan_n = 10")
plt.xlim([0, 15])
plt.ylim([-10.0, 10.0])
plt.xticks([0, 5, 10, 15])
plt.yticks([-10, -5, 0, 5, 10])
plt.grid(color='gray')
plt.xlabel('time t[s]')
plt.ylabel('y(t)')
plt.legend()
plt.show()
```
