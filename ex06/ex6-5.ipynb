```python
import numpy as np
from control import matlab
from matplotlib import pyplot as plt
from scipy import arange
```

```python
K = 1
omegan = 1
zeta1 = 1.1
zeta2 = 1.5
zeta3 = 2.0

num = [0, 0, K*omegan**2]
den1 = [1, 2*zeta1*omegan, omegan**2]
den2 = [1, 2*zeta2*omegan, omegan**2]
den3 = [1, 2*zeta3*omegan, omegan**2]

sys1 = matlab.tf(num, den1)
sys2 = matlab.tf(num, den2)
sys3 = matlab.tf(num, den3)
```

```python
t = arange(0, 10, 0.01)

y1, t1 = matlab.impulse(sys1, t)
y2, t2 = matlab.impulse(sys2, t)
y3, t3 = matlab.impulse(sys3, t)

plt.plot(t1, y1, label="zeta = 1.1")
plt.plot(t2, y2, label="zeta = 1.5")
plt.plot(t3, y3, label="zeta = 2.0")
plt.xlim([0, 10])
plt.ylim([0, 0.35])
plt.yticks([0, 0.05, 0.10, 0.15, 0.20, 0.25, 0.30, 0.35])
plt.grid(color='gray')
plt.xlabel('time t[s]')
plt.ylabel('y(t)')
plt.legend()
plt.show()
```
