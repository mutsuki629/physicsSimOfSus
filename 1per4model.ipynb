```python
# import
from control import matlab
from matplotlib import pyplot as plt
from scipy import arange
```

```python
# parameter
M1 = 400 # バネ先質量[kg]
K1 = 1400 # ばね定数[N/m]
C1 = 500 # 粘性減衰係数[Ns/m]
```

```python
# sys1
a1 = [[0, 1], [-K1/M1, 0]]
b1 = [[0], [1/M1]]
c1 = [[1, 0]]
d1 = [[0]]
sys1 = matlab.ss2tf(a1, b1, c1, d1)
print('sys1 = ', sys1)

# sys2
a2 = [[0, 1], [-K1/M1, -C1/M1]]
b2 = [[0], [1/M1]]
c2 = [[1, 0]]
d2 = [[0]]
sys2 = matlab.ss2tf(a2, b2, c2, d2)
print('sys2 = ', sys2)

# plot
t = arange(0, 10, 0.01)

y1, t1 = matlab.impulse(sys1, t)
y2, t2 = matlab.impulse(sys2, t)

plt.plot(t1, y1, label='sys1')
plt.plot(t2, y2, label='sys2')
plt.grid(color='gray')
plt.xlabel('time t[s]')
plt.ylabel('y(t)')
plt.legend()
plt.show()
```

```python
# sys3
a3 = [[0, 1], [-K1/M1, -C1/M1]]
b3 = [[0], [K1/M1]]
c3 = [[1, 0]]
d3 = [[0]]
sys3 = matlab.ss2tf(a3, b3, c3, d3)
print('sys3 = ', sys3)

# plt2
t = arange(0, 10, 0.01)
y3, t3 = matlab.impulse(sys3, t)

plt.plot(t3, y3,label='sys3')
plt.grid(color='gray')
plt.xlabel('time t[s]')
plt.ylabel('y(t)')
plt.legend()
plt.show()
```
