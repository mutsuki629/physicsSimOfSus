```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# パラメータの値を与える
a1 = -0.1 # a=-0.1の場合
a2 = -1.0 # a=-1.0の場合
a3 = -5.0 # a=-5.0の場合

# 時間変数の定義
t = np.arange(0, 10, 0.01) # 0から10まで0.01刻み

# y(t) = exp(at) の計算
y1 = np.exp(a1 * t)
y2 = np.exp(a2 * t)
y3 = np.exp(a3 * t)
```

```python
# 図1.4のプロット
plt.plot(t, y1, label="a = -0.1")
plt.plot(t, y2, label="a = -1.0")
plt.plot(t, y3, label="a = -5.0")
plt.xlim([0,10])
plt.yticks([0, 0.2, 0.4, 0.6, 0.7, 0.8, 1.0])
plt.xlabel("time t[s]")
plt.ylabel("output")
plt.legend()
plt.show()
```
