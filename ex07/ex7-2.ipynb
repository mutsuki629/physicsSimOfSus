```python
import numpy as np
from control import matlab
from matplotlib import pyplot as plt
from scipy import arange
```

```python
num = [0, 0, 1, 2]
den = [1, 2, 1, 3]

sys = matlab.tf(num, den)
```

```python
t = arange(0, 50, 0.01)

y, t = matlab.step(sys, t)

plt.plot(t, y)
plt.xlim([0, 40])
plt.ylim([-50, 40])
plt.xticks([0, 5, 10, 15, 20, 25, 30, 35, 40])
plt.yticks([-50, -40, -30, -20, -10, 0, 10, 20, 30, 40, 50])
plt.grid(color='gray')
plt.xlabel('time t[s]')
plt.ylabel('y(t)')
plt.show()
```
