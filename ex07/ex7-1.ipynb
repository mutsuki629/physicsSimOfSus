```python
import numpy as np
from control import matlab
from matplotlib import pyplot as plt
from scipy import arange
```

```python
num = [0, 0, 1, 2]
den = [1, 2, 1, 1]

sys = matlab.tf(num, den)
```

```python
t = arange(0, 50, 0.01)

y1, t1 = matlab.step(sys, t)

plt.plot(t1, y1)
plt.xlim([0, 50])
plt.ylim([0, 3.5])
plt.xticks([0, 10, 20, 30, 40, 50])
plt.yticks([0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5])
plt.grid(color='gray')
plt.xlabel('time t[s]')
plt.ylabel('y(t)')
plt.show()
```
