```python
from control import matlab
from matplotlib import pyplot as plt
from scipy import arange
```

```python
R = 100
L = 10

num = [0, 1]
den = [L, R]

sys = matlab.tf(num, den)

t = arange(0, 1.01, 0.01)

y, t = matlab.step(sys, t)

plt.plot(t, y)
plt.xlim([0, 1])
plt.yticks([0, 0.002, 0.004, 0.006, 0.008, 0.010, 0.012])
plt.grid(color='gray')
plt.xlabel('time t[s]')
plt.ylabel('i(t)[A]')
plt.show()
```
